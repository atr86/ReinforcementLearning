```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#code to plot histogram
def plot_histogram(data, xlabel_, y_label, title_):
    #plot the histogram of velocities
    plt.hist(data, bins=100, density=True)

    plt.xlabel(xlabel_, fontsize=16)
    plt.ylabel(y_label, fontsize=16)
    plt.title(title_, fontsize=16)
    plt.show()
```

```python
#sample from uniform
N = 1000_000
u = np.random.uniform(low=0., high=1., size=N)
plot_histogram(u, "u", "$f_u(u)$", "Density of U(0,1)")
```

```python
# E[u] = 0.5, \sigma = root(1/12)
mu, sigma = 0.5, np.sqrt(1/12)   # for u(0, 1)
B = 200
N2 = int(np.ceil(N/B))
norm_u = []
for i in range(N2):
    if i==N2-1:
        endpos = N
    else:
        endpos = B*(i+1)
    # rescaling -- (u - mu)/s/root(n); where u are samples from U(0,1)
    z_i = (np.mean(u[B*i:endpos]) - mu)/(sigma/np.sqrt(B))
    norm_u.append(z_i)
    
norm_u = np.array(norm_u)
plot_histogram(norm_u, "$z = \\frac{\overline{x} - \mu}{\sigma/\sqrt{N}}$", "$f_Z(z)$", "Central Limit Theorem")
```

```python
#streotypical normal density
z = np.random.normal(size=5000)
plot_histogram(z, "$z$", "$N(0,1)$", "Normal Density")
```
