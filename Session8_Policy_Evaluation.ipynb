```python
import numpy as np
```

```python
#policy_eval for equal policy
def converged(prev_state, temp, epsilon):
  for i in range(prev_state.shape[0]):
      for j in range(prev_state.shape[1]):
        if temp[i][j] == neg_inf:
          continue
        #difference between values of the current state )t=n and t=n-1 is less than some epsilon  (here the temp is the current state, and prev_state is n-1)
        if np.abs(temp[i][j] - prev_state[i][j]) > epsilon:
          return False
  return True
def policy_eval(prev_state, epsilon, terminal):
  neg_inf = -float('inf')
  temp = np.copy(prev_state)
  while True:
    for i in range(prev_state.shape[0]):  #for i = 1 to 4, j = 1 to 4
      for j in range(prev_state.shape[1]):
        if temp[i][j] == neg_inf or (i==terminal[0] and j == terminal[1]) or (i==terminal[2] and j ==terminal[3]):
          continue
        ctr_rem = 0   # says that number of times i stay back in the current state
        state_sum = 0
        #up
        if temp[i-1][j] == neg_inf:  
          ctr_rem = ctr_rem + 1
        else:
          state_sum = state_sum + temp[i-1][j]   # add the V(s')
        #left
        if temp[i][j-1] == neg_inf:
          ctr_rem = ctr_rem + 1
        else:
          state_sum = state_sum + temp[i][j-1]
        #down
        if temp[i+1][j] == neg_inf:
          ctr_rem = ctr_rem + 1
        else:
          state_sum = state_sum + temp[i+1][j]
        #right
        if temp[i][j+1] == neg_inf:
          ctr_rem = ctr_rem + 1
        else:
          state_sum = state_sum + temp[i][j+1]
        
        # r(s,a) = -1
        # pi(a|s)V(s') surrounding states
        #state sum = surrounding sum + sum the state values if you stay in it
        state_sum = ctr_rem*temp[i][j] + state_sum
        #uniform policy r=-1 on each step
        prev_state[i][j] = -1 + 0.25*state_sum
    if converged(prev_state, temp, epsilon):
      temp = np.copy(prev_state)
      break
    temp = np.copy(prev_state)
    

  return prev_state
```

```python
print("Without extra state")
neg_inf = -np.inf
prev_state2 = np.zeros((6, 6))
epsilon = 1E-10
#set the states surrounding to be 0
for i in range(prev_state2.shape[0]):
  for j in range(prev_state2.shape[1]):
    if i == 0 or j == 0 or i == prev_state2.shape[0] - 1 or j == prev_state2.shape[1]-1:
      prev_state2[i][j] = neg_inf
print(prev_state2) 
```
